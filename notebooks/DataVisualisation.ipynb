Calculate and display the correlation matrix of the dataset.

calculates the correlation matrix of the DataFrame `data` 
using the `corr()` method and stores it in the variable `corr`. The correlation 
matrix is a table showing correlation coefficients between variables. Each cell 
in the table shows the correlation between two variables. The value is in the 
range of -1 to 1, where 1 means highly correlated, 0 means no correlation, and 
-1 means highly negatively correlated.

Returns:
    DataFrame: A correlation matrix of the dataset.

### count of cars by car make

```python
data['manufacturer'] = data['manufacturer'].astype(str).str.strip()
fig = plt.figure(figsize=(20, 15))
sns.countplot(y='manufacturer', data=data, order=data['manufacturer'].value_counts().index)
plt.title("Distribution of car manufacturers")
plt.xticks(rotation=90)
plt.savefig('../images/distribution-of-car-manufacturers.png')
```

### create a pivot table to find the average price per car manufacturer

This scatter plot shows the relationship between the age of cars and their prices. Each point represents a car, with the color indicating the type of engine. Generally, there is a negative correlation between the age and price of cars, meaning that as cars get older, their prices tend to decrease.

```python
engine_colors = {
    'Petrol': 'blue',
    'Diesel': 'green',
    'Hybrid': 'orange',
    'Electric': 'red',
    'Plug_in_hybrid': 'purple'
}
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='price', data=data, hue='engine', palette=engine_colors, alpha=0.7, edgecolor=None)
plt.title('Scatter Plot of Car Age vs Price', fontsize=16)
plt.xlabel('Age of Car (years)', fontsize=14)
plt.ylabel('Price of Car (£)', fontsize=14)
plt.legend(title='Engine Type')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('../images/cars_age_vs_price.png')
plt.show()
```

There is negative correlation between the age and prices of cars. As the age increases, the price reduces

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='kilometerage', y='price', data=data, hue='engine', palette=engine_colors, alpha=0.5, edgecolor=None, s=20)
plt.title('Scatter Plot of Kilometerage vs Price', fontsize=16)
plt.xlabel('Kilometerage', fontsize=14)
plt.ylabel('Price of Car (£)', fontsize=14)
plt.legend(title='Engine Type')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('../images/kilometerage_vs_price.png')
plt.show()
```

There is negative correlation between the mileage and prices of the cars. As the number of miles travelled increases, the price reduces

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='kilometerage', data=data, hue='engine', palette=engine_colors, alpha=0.7, edgecolor=None)
plt.title('Scatter Plot of Car Age vs Kilometerage', fontsize=16)
plt.xlabel('Age of Car (years)', fontsize=14)
plt.ylabel('Kilometerage', fontsize=14)
plt.legend(title='Engine Type')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('../images/age_vs_kilometerage.png')
plt.show()
```

There is a positive correlation between the mileage and age of the cars. the number of miles travelled increases with age.

## 3D bar chart for manufacturer frequency


```python
from matplotlib import cm
summary_stats = data.describe()
print(summary_stats)

fig = plt.figure(figsize=(25, 20))
ax = fig.add_subplot(111, projection='3d')
manufacturer_counts = data["manufacturer"].value_counts()
x = np.arange(len(manufacturer_counts))
y = manufacturer_counts.values
z = np.zeros_like(x)

ax.bar(x, y, zs=z, zdir='y', alpha=0.8, color=cm.viridis(np.linspace(0, 1, len(x))))
ax.set_xticks(x)
ax.set_xticklabels(manufacturer_counts.index, rotation=45, ha="right")
ax.set_title("3D Bar Chart of Manufacturer Frequency")
ax.set_ylabel("Frequency")
ax.set_zlabel("Count")
plt.tight_layout()
plt.savefig("../images/visualizations_output.png")
plt.show()
```

## Visualization: Most common car names

```python
plt.figure(figsize=(15, 8))
car_name_counts = data["name"].value_counts().head(40)
sns.barplot(y=car_name_counts.index, x=car_name_counts.values, palette="coolwarm")
plt.title("Top 20 Most Common Car Names", fontsize=16)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Car Name", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("../images/car_names_visualization_head.png")
plt.show()
```

## Visualization: Least common car names


```python
plt.figure(figsize=(15, 8))
car_name_counts = data["name"].value_counts().tail(40)
sns.barplot(y=car_name_counts.index, x=car_name_counts.values, palette="coolwarm")
plt.title("Last 20 Least Common Car Names", fontsize=16)
plt.xlabel("Count", fontsize=14)
plt.ylabel("Car Name", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("../images/car_names_visualization_tail.png")
plt.show()
```
